# 3-Data Processing

```python
data=data.drop(['ID','Doors'],axis=1)
data.head()
```

# Date

```python
data['Age']=dtime.year-data['Prodyear']
```

```python
data=data.drop(['Prodyear'],axis=1)
```

# Levy

```python
data.Levy.replace({'-':'0'},inplace=True)
data['Levy']=data['Levy'].astype(int)
```

```python
data.head()

    
```

## Mileage

```python
data['Mileage']=data['Mileage'].str.replace('km','')
```

```python
data.head()
```

## Engine volume

```python
data['Engine volume']=data['Engine volume'].str.replace('Turbo','')
data['Engine volume']=data['Engine volume'].astype(float)
```

```python
data.head()
```

```python
data.info()
```

## Outliers 

```python
data_num=data.select_dtypes(exclude='object')
for col in data_num:
    q1=data[col].quantile(0.25)
    q3=data[col].quantile(0.75)
    iqr=q3-q1
    low=q1-1.5*iqr
    high=q3+1.5*iqr
    outlier=((data_num[col]>high)|(data_num[col]<low)) .sum()
    total=data_num[col].shape[0] 
    print(f"total outliers in {col} are: {outlier}--{round(100*outlier/total,2)}%")
    if outlier>0:
        data=data.loc[(data[col]<=high) & (data[col]>=low)]
```

## Transform Data

```python
dobject=data.select_dtypes(include='object')
dnum=data.select_dtypes(exclude='object')
```

```python
la=LabelEncoder()
```

```python
for i in range(0,dobject.shape[1]):
   dobject.iloc[:,i]= la.fit_transform(dobject.iloc[:,i])
```

```python
data=pd.concat([dobject,dnum],axis=1)
data
```

```python
data.info()
```

```python
for col in data.columns:
    if data[col].dtype == 'object':
        try:
            data[col] = pd.to_numeric(data[col])
        except ValueError:
            pass  # If conversion fails, leave the column as object
print(data.dtypes)
```

# 4-Model

```python
x=data.drop('Price',axis=1)
y=data['Price']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
Algorithms=['LinearRegression','DecisionTreeRegressor','RandomForestRegressor','GradientBoostingClassifier','XGBRegressor','SVR']
R2=[]
RMSE=[]
```

```python
def models(model):
    model.fit(x_train,y_train)
    pre=model.predict(x_test)
    r2=r2_score(y_test,pre)
    R2.append(r2)
    rmse=np.sqrt(mean_squared_error(y_test,pre))
    RMSE.append(rmse)
    score=model.score(x_test,y_test)
    print(f"the score of the model is : {score}")
    
```

```python
model1=LinearRegression()
model2=DecisionTreeRegressor(random_state=42)
model3=RandomForestRegressor()
model4=GradientBoostingRegressor()
model5=XGBRegressor()
model6=SVR()
```

```python
models(model1)
models(model2)
models(model3)
models(model4)
models(model5)
models(model6)
```

```python
df=pd.DataFrame({'Algorithms':Algorithms,'R2_score':R2,'RMSE':RMSE})
df
```

```python
fig,sx=plt.subplots(figsize=(20,5))
plt.plot(df.Algorithms,df.R2_score,label='R2_score',marker='x')
plt.legend()
plt.show()
```

```python
fig,sx=plt.subplots(figsize=(20,5))
plt.plot(df.Algorithms,df.RMSE,label='RMSE',c='r',marker='o')
plt.legend()
plt.show()
```
